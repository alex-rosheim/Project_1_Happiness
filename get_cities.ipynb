```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
cbsa_cities = pd.read_csv("cbsa_cities.csv", encoding = 'latin-1')
top_25 = pd.read_csv('top_25_key.csv')
bottom_50 = pd.read_csv('miserable_cities.csv')
```

```python
cbsa_metros = cbsa_cities[cbsa_cities['Metropolitan/Micropolitan Statistical Area'] == 'Metropolitan Statistical Area']
len(cbsa_metros)
```

```python
fips_codes = pd.read_csv("fips_codes.csv")
fips_codes.columns = ['FIPS State Code', 'STUSAB', 'State Name', 'STATENS']
```

```python
cities_df = pd.merge(cbsa_cities, fips_codes, on='FIPS State Code')
cities_df['city_state'] = cities_df['Principal City Name'] + ", " + cities_df['STUSAB']
cities_df['city_full_state'] = cities_df['Principal City Name'] + ", " + cities_df['State Name']
cities_df.head()
```

```python
cities_list = cities_df['city_state'].tolist()
bottom_list = bottom_50['City_State'].tolist()
miserable_list = cities_df['city_state'].isin(bottom_list)
cities_df['Miserable City'] = miserable_list
```

```python
cities_list = cities_df['city_full_state'].tolist()
top_list = top_25['metro'].tolist()
happy_list = cities_df['city_full_state'].isin(top_list)
cities_df['Happy City'] = happy_list
```
